```python
category_counts = user_reviews['App'].value_counts()
category_counts
```

```python
#MERGING THE DATASETS
merged_data = pd.merge(apps, user_reviews, on='App', how='inner')
```

```python
merged_data
```

```python
#METRICS ANALYSIS 3
ratings_stats = apps['Rating'].describe()
size_stats = apps['Size'].describe()
installs_stats = apps['Installs'].describe()
print("Rating Statistics:")
print(ratings_stats)

print("\nSize Statistics:")
print(size_stats)

print("\nInstalls Statistics:")
print(installs_stats)
```

```python
#SENTIMENT ANALYSIS 4

# Count the sentiment with the highest occurrence
most_common_sentiment = merged_data['Sentiment'].value_counts().idxmax()

# Get the highest count sentiment
most_common_count = merged_data['Sentiment'].value_counts().max()


print(f"The most common sentiment is '{most_common_sentiment}' with a count of {most_common_count}.")

```

```python
#VISUALISATION : 1
import matplotlib.pyplot as plt

# Creating a pie chart for sentiment distribution
sentiment_distribution = merged_data['Sentiment'].value_counts()
labels = sentiment_distribution.index
counts = sentiment_distribution.values

colors = ['green', 'red', 'blue']

plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Sentiment Distribution')
plt.show()
```

```python
#VISUALISATION : 2
import matplotlib.pyplot as plt

# Plotting the boxplot
merged_data.groupby('Category')['Sentiment'].value_counts(normalize=True).unstack().plot(kind='box', stacked=True)
plt.title('Sentiment Distribution by App Category')
plt.xlabel('App Category')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

INSIGHTS FROM THIS VISUALISATION:

---
The sentiment distribution by app category explains which app has more positive feedback percentage and which has negetive feedback percentage. The categories are rated with the help of sentiments which are categorised in "Positive", "Negetive", "Neutral".
As we can see that the apps in google playstore, they are given more positive feedback and medium negetive feedback and less neutral feedback.
With the help of outliers in these category what we can understand is,

*   "Negative" category contains one category present which has no negative feedback. It has only "Positive" and "Neutral" feedback.
*   "Neutral" category contains three categories which has no neutral feedback. Its has either "Positive" or "Negative" feedback present.
*   And the "Positive" category has no outliers which means that every category has poitive ratings and feedback.






```python
#Visualisation : 3
import matplotlib.pyplot as plt

# Assuming 'merged_data' is your merged DataFrame

# Creating a line chart
sentiment_counts = merged_data.groupby(['Rating', 'Sentiment']).size().unstack(fill_value=0)
sentiment_counts.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Distribution of Sentiments Across App Ratings')
plt.xlabel('App Ratings')
plt.ylabel('Number of Reviews')
plt.legend(title='Sentiment', loc='upper right', labels=['Negative', 'Neutral', 'Positive'])
plt.show()
```

INSIGHTS FROM THIS VISUALISATION OF LINE CHART


---
As shown above, the relation between the reviews and ratings of the app.
we can see that, the less the ratings are the correlation wiht the reviews is also less.
As the app rating increases at 3.5 the reviews starts to rise with a slow start and has the highest reviews when the ratings are between 4.0 to 4.5.


```python
#VISUALISATION : 4
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'merged_data' is your merged DataFrame

# Convert 'Installs' column to numeric (remove '+' and ',' characters)
merged_data['Installs'] = merged_data['Installs'].str.replace(',', '').str.replace('+', '').astype(int)

# Calculate the total installs for each category
category_installs = merged_data.groupby('Category')['Installs'].sum().sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
category_installs.plot(kind='bar', color='skyblue')
plt.title('Total Installs by App Category')
plt.xlabel('App Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
